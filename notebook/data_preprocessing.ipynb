### Replace 'Anomaly Data' with NaN in the User_Score column

```python
df['User_Score'] = df['User_Score'].replace('tbd', np.nan)
```

### Checking Unique Data

```python
for kolom in df.columns:
    uniques = df[kolom].unique()
    print(f"Data unik di kolom {kolom} ({len(uniques)}):")
    print(list(uniques))
    print('-' * 40)
```

### Converting the data types of columns

```python
df['User_Score'] = pd.to_numeric(df['User_Score'], errors='coerce')
df['Year_of_Release'] = pd.to_numeric(df['Year_of_Release'], errors='coerce')
```

```python
df = df.dropna(subset=['Year_of_Release'])
df['Year_of_Release'] = df['Year_of_Release'].astype(int)
```

```python
selected_columns = [
    'Name',
    'Platform',
    'Year_of_Release',
    'Genre',
    'Publisher',
    'NA_Sales',
    'EU_Sales',
    'JP_Sales',
    'Other_Sales',
    'Global_Sales',
    'Critic_Score',
    'User_Score'
]

df = df[selected_columns]

print(f"Dimensi Akhir : {df.shape}")
print(f"Daftar Kolom : {df.columns.to_list()}")
```

### Strategic Filtering (Business Logic)

#### **Objective:** Isolate relevant market data by time and platform technology.

```python
platform_years = df.groupby('Platform')['Year_of_Release'].unique()
print(platform_years)
```

```python
df = df[df['Year_of_Release'] >= 2013]
zombie_platforms = ['DS', 'Wii', 'PSP', 'PS2']
df = df[~df['Platform'].isin(zombie_platforms)]
```

```python
df.info()
```

### Save Clean Data to CSV File

```python
import os

output_folder = '../dataset'

os.makedirs(output_folder, exist_ok=True)

file_path = os.path.join(output_folder, 'clean_data_video_games.csv')

df.to_csv(file_path, index=False)

print(f"File berhasil disimpan di: {file_path}")
```
