### Importing Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

### Importing Dataset

```python
df = pd.read_csv("..\dataset\clean_data_video_games.csv")
```

```python
df
```

### Clean Data Visualization

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['Global_Sales'], bins=30, kde=True, color='green')
plt.title(f'Distribusi Penjualan (Skewness: {df["Global_Sales"].skew():.2f})')
plt.xlabel('Global Sales')
```

```python
plt.subplot(1, 2, 2)
sns.histplot(df['Critic_Score'], bins=20, kde=True, color='blue')
plt.title(f'Distribusi Critic Score (Skewness: {df["Critic_Score"].skew():.2f})')
plt.xlabel('Score')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.countplot(data=df, x='Platform', order=df['Platform'].value_counts().index, palette='viridis')
plt.title('Jumlah Game per Platform (Sample Size)')
```

```python
plt.subplot(1, 2, 2)
sns.countplot(data=df, y='Genre', order=df['Genre'].value_counts().index, palette='magma')
plt.title('Jumlah Game per Genre')
plt.tight_layout()
plt.show()
```

```python
corr_matrix = df[['Global_Sales', 'Critic_Score', 'User_Score']].corr()

# Visualisasi Heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Korelasi: Sales vs Scores')
plt.show()

print("Insight Bisnis:")
print(f"- Korelasi Critic Score vs Global Sales: {corr_matrix.loc['Global_Sales', 'Critic_Score']:.2f}")
```
